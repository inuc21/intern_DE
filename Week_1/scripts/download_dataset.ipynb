## About Dataset
E-Commerce Sales Analytics (5,000 Records)
Dataset Description
Context
In the highly competitive digital retail landscape, understanding customer behavior, purchasing patterns, and sales performance is crucial for driving growth. This dataset contains 5,000 synthetic yet highly realistic retail transactions designed to mirror modern online marketplace dynamics. It bridges customer demographics, product categorization, financial metrics, and operational logistics into a single, cohesive analytics asset.

Content & Structure
The dataset consists of a single CSV file (ecommerce_sales_analytics_5000.csv) containing 5,000 unique rows and 15 analytical columns. It tracks the end-to-end lifecycle of retail orders, allowing data scientists to practice everything from basic data cleaning to complex machine learning modeling.

Key Features:
Demographics: Customer Age, Gender, and Location.

Financials: Unit Prices, Quantities, applied Discount Percentages, and Net Revenue.

Logistics: Time deltas between Order Dates and Shipping Dates.

```python
import kagglehub
import shutil
from pathlib import Path

# Download dataset
download_path = kagglehub.dataset_download(
    "abbas829/ecommerce-sales-dataset"
)

print("Downloaded to:", download_path)

# Thư mục đích trong project
target_dir = Path.home() / "intern_DE" / "Week_1" / "data" / "raw"
target_dir.mkdir(parents=True, exist_ok=True)

# Copy file sang raw/
for file in Path(download_path).iterdir():
    if file.is_file():
        shutil.copy(file, target_dir / file.name)

print(f"Copied dataset to: {target_dir}")
```
